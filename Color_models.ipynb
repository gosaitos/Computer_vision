### RGB TO CMY CONVERSION: ###

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
img = cv2.cvtColor(cv2.imread("pg.jpg"),cv2.COLOR_BGR2RGB)
img.shape
```

```python
img_cmy = 255 - img
```

```python
plt.figure(figsize=(12,12))
plt.subplot(1,2,1)
plt.imshow(img)
plt.axis('off')
plt.title('RGB image')
plt.subplot(1,2,2)
plt.imshow(img_cmy)
plt.axis('off')
plt.title('CMY image')
plt.show()
```

### RGB TO HSV ###

```python
img = cv2.imread("rdr2.webp")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)
h, s, v = cv2.split(img_hsv)

h += 3
s += 50
v //= 2

img_adj = cv2.merge([h,s,v])
imgf = cv2.cvtColor(img_adj, cv2.COLOR_HSV2RGB)
plt.figure(figsize=(10,5))

def plot(x,y,w,z):
    plt.subplot(2, 3, z)
    plt.imshow(x,cmap=y)
    plt.title(w)
    plt.axis('off')

plot(h,'gray','Hue Channel',1)
plot(s,'gray','Saturation Channel' , 2)
plot(v,'gray','Value Channel' , 3)
plot(img_rgb, None,'RGB image' , 4)
plot(img_hsv,None,'HSV image' , 5)
plot(imgf,None,'Final image' , 6)
plt.tight_layout()
plt.show()
```

### RGB SPLIT ###

```python
img1 = cv2.cvtColor(cv2.imread("rdr2.webp"),cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,5))
def rgb(i,j,k,l):
    img = img1.copy()
    if i is not None and j is not None:
        img[:,:,i]= 0
        img[:,:,j]= 0
    plt.subplot(2,2,l)
    plt.imshow(img)
    plt.title(f"{k} image")
    plt.axis('off')

rgb(None,None,'RGB',1)
rgb(1,2,'R - Channel',2)
rgb(0,2,'G - Channel',3)
rgb(0,1,'B - Channel',4)
plt.tight_layout()
plt.show()
```

### CMY SPLIT ###

```python
img_rgb = cv2.cvtColor(cv2.imread("rdr2.webp"),cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,5))

def cmy(i, k, l):
    cmy = img_rgb.copy()
    if i is not None:
        cmy[:, :, i] = 0
    plt.subplot(2, 2, l)
    plt.imshow(cmy)
    plt.title(f"{k} image")
    plt.axis('off')

cmy(None, 'RGB', 1)
cmy(0, 'Cyan', 2)
cmy(1, 'Magenta', 3)
cmy(2, 'Yellow', 4)

plt.tight_layout()
plt.show()
```

### RGB TO CMY USING OWN IMG:

```python
import numpy as np
import matplotlib.pyplot as plt

rgb_image = np.array([
    [[255, 0, 0], [0, 255, 0], [0, 0, 255], [255, 255, 0]],
    [[0, 255, 255], [255, 0, 255], [128, 128, 128], [255, 255, 255]],
    [[0, 0, 0], [128, 0, 0], [0, 128, 0], [0, 0, 128]],
    [[128, 128, 0], [0, 128, 128], [128, 0, 128], [64, 64, 64]]
], dtype=np.uint8)

cmy_image = 255 - rgb_image

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.imshow(rgb_image)
plt.title('RGB Image')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(cmy_image)
plt.title('CMY Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
